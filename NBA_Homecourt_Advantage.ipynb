## Bar Plot indicating win percentage for Home and Away games

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

# Sort Win% at home by descending order
frame_home_2014= frame_home_2014.sort_values("W_PCT", ascending=False)

# Plot the Win% at home games
sns.set_color_codes("pastel")
sns.barplot(x=frame_home_2014["W_PCT"], y=frame_home_2014["TEAM_NAME"], data=frame_home_2014,label="Home", color="r")

# Sort Win% away by descending order
frame_away_2014= frame_away_2014.sort_values("W_PCT", ascending=False)

# Plot the Win% at away games
sns.set_color_codes("muted")
sns.barplot(x="W_PCT", y="TEAM_NAME", data=frame_away_2014,label="Away", color="r")

#Add a legend and informative axis label
plt.suptitle("NBA Team's Win Percentage", y=0.90)
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",
       xlabel="Win Percentage")
sns.despine(left=True, bottom=True)
```

## Scatter Plot depicting Home and Away win% trend

```python

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 8))
plt.suptitle("Win Percentage vs Field Goal Percentage", y=0.93)
sns.scatterplot(data=frame_home_2014, x=frame_home_2014["W_PCT"], y=frame_home_2014["FG_PCT"], color="b", label="Home")
sns.scatterplot(data=frame_away_2014, x=frame_away_2014["W_PCT"], y=frame_away_2014["FG_PCT"], color="r", label="Away")
ax.set(ylabel="Win Percentage",
       xlabel="Field Goal Percentage")
sns.despine(left=True, bottom=True)
ax.legend(ncol=2, loc="lower right", frameon=True)
#set a border around plot
ax.patch.set_edgecolor('black')  
ax.patch.set_linewidth('1')
# Hide grid lines
ax.grid(False)
```

# Calculating Mean and Median win percentages for Home and Away games

```python
#mean win% of all teams for the 2014/15 NBA season, Home and Away
import statistics
#home mean
W_PCT_home = frame_home_2014["W_PCT"]
win_home=statistics.mean(W_PCT_home)
print(win_home)
#away mean
W_PCT_away = frame_away_2014["W_PCT"]
win_away=statistics.mean(W_PCT_away)
print(win_away)
```

Calculating how many more games team win at home vs away

```python
((win_home-win_away)/win_away)*100
```
